# Data Analysis on Amazon Prime Video dataset

```python
#import all Libraries
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
prime=pd.read_csv("Data//amazon.csv")#Read CSV file
```

## first we will see no. of movies per year from graph

```python
plt.figure(figsize=(15,10))# it shows graph size
prime['Year of release'].value_counts().plot(kind="bar") # pandas plot
plt.title("Movies per year")
plt.ylabel("No. Of Movies")
plt.xlabel("Year of release")
plt.xticks(rotation=90)
plt.show()
```

## Now we will see no. of movies per language from 

```python
plt.figure(figsize=(15,10))
prime['Language'].value_counts().plot(kind="bar") # pandas plot
plt.title("Language wise Movies")
plt.ylabel("No. Of Movies")
plt.xlabel("Language")
plt.xticks(rotation=90)
plt.show()
```

## Now we will see Genre wise movies

```python
plt.figure(figsize=(15,10))
prime['Genre'].value_counts().plot(kind="barh") # pandas plot
plt.title("Genre wise Movies")
plt.ylabel("Genre")
plt.xlabel("No. of Movies")
plt.xticks(rotation=90)
plt.show()
```

## Rating wise Movies

```python
plt.figure(figsize=(15,10))
prime['IMDb rating'].value_counts().plot(kind="bar") # pandas plot
plt.title("Rating wise Movies")
plt.ylabel("No. Of Movies")
plt.xlabel("IMDb rating")
plt.xticks(rotation=90)
plt.show()
```

## Now we will see Age Restriction on movies

```python
#here we will show pie plot
plt.figure(figsize=(7,7))
prime["Age of viewers"].value_counts().plot.pie(autopct="%1.2f%%")
plt.title("Age of Viewers")
plt.show()
```
